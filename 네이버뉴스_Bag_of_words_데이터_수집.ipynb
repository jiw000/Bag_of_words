#### 경제 뉴스

```python
economy = []

categories = [
("259", "금융"),
("258","증권"),
("261","산업/재계"),
("771","중기/벤처"),
("260","부동산"),
("262","글로벌경제"),
("310","생활경제"),
("263","경제일반")
]

classes = ["type06_headline", "type06"]
date = "20230221"


for cate_code, cate_name in categories: 

    url = f"https://news.naver.com/main/list.naver?mode=LS2D&mid=shm&sid2={cate_code}&sid1=101&date={date}&page=999"
    headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36"}
    html = requests.get(url, headers=headers).text
    soup = BeautifulSoup(html, "html.parser")
    max_page = soup.select("#main_content > div.paging > strong")[0].text
    max_page = int(max_page)

    for i in tqdm(range(1,max_page)):
        url = f"https://news.naver.com/main/list.naver?mode=LS2D&mid=shm&sid2={cate_code}&sid1=101&date={date}&page={i}"
        headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36"}
        html = requests.get(url, headers=headers).text
        soup = BeautifulSoup(html, "html.parser")

        for c in classes:
            for n in range(1,11):
                string = f"#main_content > div.list_body.newsflash_body > ul.{c} > li:nth-child({n}) > dl > dt:nth-child(2) > a"
                temp = soup.select(string)
                for line in temp:
                    res_detail = {}
                    res_detail['name'] = line.text.strip()
                    res_detail['link'] = line['href']
                    res_detail["category"] = cate_name
                    res_detail["section"] = "경제"
                    economy.append(res_detail)


economy
```

```python
for single in tqdm(economy):

    link = single["link"]

    url = link
    
    headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36"}
    html = requests.get(url, headers=headers).text
    soup = BeautifulSoup(html, "html.parser")

    string = "#dic_area" 
    temp = soup.select(string)

    if temp:
        single['text'] = temp[0].text.strip()
    else:
        single["text"] = None
                                     
```

```python
df = pd.DataFrame(economy)
df.to_csv("economy_news")
df
```

### 네이버 뉴스기사 수집 함수

```python
def naver_news_href(categories, date, section, section_name):

    news = []

    classes = ["type06_headline", "type06"]
    date = date


    for cate_code, cate_name in categories: 
        url = f"https://news.naver.com/main/list.naver?mode=LS2D&mid=shm&sid2={cate_code}&sid1={section}&date={date}&page=999"
        headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36"}
        html = requests.get(url, headers=headers).text
        soup = BeautifulSoup(html, "html.parser")
        max_page = soup.select("#main_content > div.paging > strong")[0].text
        max_page = int(max_page)
        
        for i in tqdm(range(1,max_page)):
            url = f"https://news.naver.com/main/list.naver?mode=LS2D&mid=shm&sid2={cate_code}&sid1={section}&date={date}&page={i}"
            headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36"}
            html = requests.get(url, headers=headers).text
            soup = BeautifulSoup(html, "html.parser")

            for c in classes:
                for n in range(1,11):
                    string = f"#main_content > div.list_body.newsflash_body > ul.{c} > li:nth-child({n}) > dl > dt:nth-child(2) > a"
                    temp = soup.select(string)
                    for line in temp:
                        res_detail = {}
                        res_detail['name'] = line.text.strip()
                        res_detail['link'] = line['href']
                        res_detail["category"] = cate_name
                        res_detail["section"] = section_name
                        news.append(res_detail)

    return news
```

```python
# test 
categories = [
("731", "모바일"),
("226","인터넷/SNS")
]

test = naver_news_href(categories,"20230221",101,"IT")
```

```python
def naver_news_text(href):
    for single in tqdm(href):

        link = single["link"]

        url = link
        
        headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36"}
        html = requests.get(url, headers=headers).text
        soup = BeautifulSoup(html, "html.parser")

        string = "#dic_area" 
        temp = soup.select(string)

        if temp:
            single['text'] = temp[0].text.strip()
        else:
            single["text"] = None
```
