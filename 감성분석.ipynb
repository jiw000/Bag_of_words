```python
import nltk
from konlpy.tag import Okt
from tqdm import tqdm
okt = Okt()
```

```python
pos_map = {}
for i in tqdm(data["문장"].index):
    text = data.loc[i, "문장"]
    poss = okt.pos(text,stem = True)
    pos_map[i] = poss

pos_map
```

```python
aavn = {}

for i in tqdm(data['문장'].index):
    text = pos_map[i]

    aavn[i] = []
    for word, tag in text:
        if tag in ['Noun','Adjective','Verb','Adverb']:
            aavn[i].append(word)

aavn
```

```python
# 단어 중복제거

unique_words = []

for words in tqdm(aavn.values()):
    unique_words.extend(words)
    unique_words = list(set(unique_words))
```

```python
len(unique_words)
```

```python
count = pd.DataFrame(columns = unique_words)
count
```

```python
Data = []
for i in tqdm(data['문장'].index):

    vc = pd.Series(aavn[i]).value_counts()
    datas = vc.to_dict()
    Data.append(datas)
    
count_df = pd.DataFrame(Data, index = data.index)

count_df
```

```python
count_df = count_df.fillna(0)
```

```python
data['감정']
```

```python
count_df = pd.concat([count_df,data['감정']], axis =1)
count_df
```

```python
unique_key = data["감정"].unique()

section_map = {}
section_unmap = {}
for i, feeling in enumerate(unique_key):
    section_map[feeling] = i
    section_unmap[i] = feeling

section_map
```

```python
count_df["감정"] = count_df["감정"].replace(section_map)
count_df
```

```python
feeling = count_df['감정'].iloc[:, 1:]
feeling
```

```python
count_Df = count_df.drop('감정', axis = 1)
count_Df
```

```python
count_Df['감정'] = feeling
count_Df
```

```python
count_Df['감정']
```

### 감정 예측(Random Forest)

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
# 변수 설정
feature = list(count_Df.columns.difference(['감정']))
x = count_Df[feature]
y = count_Df['감정']

# 학습데이터와 평가데이터의 비율을 8:2 로 분할
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size = 0.2, random_state = 42)
print(train_x.shape, test_x.shape, train_y.shape, test_y.shape) 
```

```python
clf = RandomForestClassifier(n_estimators=100,random_state=0)
clf.fit(train_x,train_y)

predict = clf.predict(test_x)
print(accuracy_score(test_y, predict))
```

```python
import numpy as np

importances = clf.feature_importances_
std = np.std([tree.feature_importances_ for tree in clf.estimators_], axis=0)
indices = np.argsort(importances)[::-1]

print("Feature ranking:")

for f in range(x.shape[1]):
    print("{}. feature {} ({:.4f})".format(f + 1, x.columns[indices][f], importances[indices[f]]))
```
